## Problem 1: Smart Light Controller
**Problem Statement**

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
## Smart Light Controller

## Creating the Light class
class Light:
    ## Constructor
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   ## Default value

    ## Method to turn on the lights
    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is turned {self.status}")

    ## Method to turn off the lights
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is turned {self.status}")

    ## Method to display the Light status
    def display(self):
        print(f"{self.light_name} is {self.status}")
```

```python

## Creating the object named bedroom_light
bedroom_light = Light("Bedroom Light")

## Calling the method to turn on the light
bedroom_light.turn_on()

## Calling the method to Display the light status
bedroom_light.display()
```

## Problem 2: Employee ID Card System
**Problem Statement**

Create a class that stores employee name, ID, and department and displays ID card details.


```python
## Employee ID Card System

## Creating Employee class
class Employee:

    ## Constructor to store employee details
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    ## Method to display Employee details
    def display(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.dept}")
```

```python

## Creating the object
emp_1 = Employee("Rahul", "EMP102", "IT")

## Calling the display method to display the employee details
emp_1.display()
```

## Problem 3: Mobile Contact Record
**Problem Statement**

Create a class to store contact name and phone number and display contact information.

```python
## Mobile Contact Record

## Creating a class that can store and display contact details
class Contacts:

    ## Creating the constructor to store contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
        ## prints "contact saved" whenever an new object is created
        print("Contact Saved")

    ## Method to display contact details
    def display(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
```

```python
## Creating object
anita = Contacts("Anita",  9876543210)
## Calling the display() method to display contact details
anita.display()
```

## Problem 4: Product Price Tag Generator
**Problem Statement**

Create a class that stores product name and price and prints a formatted price tag.


```python
## Product Price Tag Generator

## Creating Products Class
class Products:

    ## Creating constructor to store product_name and product_price
    def __init__(self, name, price):
        self.name = name
        self.price = price

    ## Method to add Price Tag and to display product_name and product_price
    def display(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
```

```python
## Creating object "Headphones"

Headphones = Products("Headphones", 2499)

## Calling the display method
Headphones.display()
```

## Problem 5: Movie Rating Display System
**Problem Statement**

Create a class that stores movie name and rating and displays movie details.


```python
## Creating class "Ratings" to store and display movie ratings
class Ratings:

    ## Creating constructor to store mavie name and rating
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    ## Method to display movie name and mavie rating
    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
```

```python
## Creating object "Inception"

inception = Ratings("Inception", 4.8)

## Calling the method to display movie name and rating
inception.display()
```

## Problem 6: Delivery Address Manager
**Problem Statement**

Create a class that stores customer name and delivery address and prints delivery details.


```python
## Delivery Address Manager

## Creating Address class to store customer details
class Address:

    ## Constructor that stores customer name, customer address
    def __init__(self, name, place):
        self.name = name
        self.place = place

    ## Method that displays Customer name and Customer Address
    def display(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.place}")
```

```python
## Creating a Customer1 object
customer1 = Address("Suman", "Hyderabad")
## Calling the method to display customer details
customer1.display()
```
