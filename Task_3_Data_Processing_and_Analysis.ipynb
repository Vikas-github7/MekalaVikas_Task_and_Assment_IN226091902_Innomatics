## Problem Statement 1: Employee Performance Bonus Eligibility

**Description:**

A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.

**Requirements:**
- Identify the highest performance score.
- Handle ties if multiple employees have the same highest score.
- Display all employees eligible for the top performance bonus.


```python
## Take the number of employees from the user
no_of_employees = int(input("Enter number of employees: "))

## Empty dictionary
performance = {}
## Logic to get the employee name and employee score from the user
for i in range(no_of_employees):
    ## Getting the employee name
    key = input("Enter the employee name: ")
    ## Getting the employee performance score
    value = int(input("Enter the employee performance score: "))
    ## Storing the employee name and performance score as dictionary

    performance[key] = value

## Logic to find highest performance score
## To store the Employees eligible for bonus 
highest_performances = []

## Highest performance score
highest_score = max(performance.values())

for i in performance:
    if performance[i] == highest_score:
        highest_performances.append(i)

print(f"Top performers Eligible for bonus:  {", ".join(highest_performances)} (Score: {highest_score})")
    
```

## Problem Statement 2: Search Query Keyword Analysis
**Description:**

An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.

**Requirements:**

- Convert the input to lowercase.
- Ignore common punctuation.
- Count the frequency of each keyword.
- Display only keywords searched more than once.


```python
## User Input and converting into lowe case

search = input("Enter the search query: ").lower()

## Splitting the search queary into words
search = search.strip('"')
words = search.split()

## initializing an empty dictionary to store frequent key words
frequent = {}
## General punctuations
punctuations = [",", ".", "!", "'",";","?",'''"''']

## Logic to get key words that occured more than once in a query
for i in words:
    if i in frequent:
        continue
    else:
        if i in punctuations:
            continue
        else:
            count = 0
            for j in words:
                if i == j:
                    count += 1
            if count > 1:
                frequent[i] = count
print(frequent)
```

## Problem Statement 3: Sensor Data Validation
**Description:**

A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.

**Requirements:**

- Identify readings that are even numbers (valid readings).
- Store them as (hour_index, reading_value) pairs.
- Ignore odd readings (invalid readings).


```python
## Taking number of readings from the user
no_of_readings = int(input("Enter number of readings: "))

## Empty list to store the sensor readings
readings = []

## Logic to get the sensor readings from the user
for i in range(no_of_readings):
    read = int(input("Enter the readings: "))
    readings.append(read)
print("Valid Sensor Readings (Hour, Value):")
## printing the list of valid sensor readings using list comprehensions
print([(i,r) for i,r in enumerate(readings) if r%2 == 0])
```

## Problem Statement 4: Email Domain Usage Analysis
**Description:**

A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.

**Requirements:**

- Count how many users belong to each email domain.
- Calculate the percentage usage of each domain.


```python
## Taking the number of emails from the user
no_of_emails = int(input("Enter number of emails: "))

## Empty list to store emails
emails = []

## Initilizing the counts of ech domain
gmail_count = 0
yahoo_count = 0

## Logic to get emails from the user
for i in range(no_of_emails):
    emails.append(input("Enter email Id: "))

## Logic to count the number of users belongs to each domain    
for i in emails:
    if "gmail" in i:
        gmail_count += 1
    elif "yahoo" in i:
        yahoo_count += 1

print(f"gmail.com: {gmail_count/len(emails) * 100}%")
print(f"yahoo.com: {yahoo_count/len(emails)*100}%")
```

## Problem Statement 5: Sales Spike Detection
**Description:**

A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.

**Requirements:**

- Calculate the average daily sales.
- Detect days where sales are more than 30% above average.
- Display the day number and sale value.


```python
## Taking number of days from the user
no_of_days = int(input("Enter number of days: "))

## Empty list to store sales data
sales = []

## Logic to take sales data from the user

for i in range(no_of_days):
    sales.append(int(input("Enter sales data: ")))

## Average daily sales 
avg_sales = sum(sales)/len(sales)

## Logic to detect unusual sales

for i, sale in enumerate(sales):
    if sale >= 1.3 * (avg_sales):
        print(f"Day {i}: {sale}")
```

## Problem Statement 6: Duplicate User ID Detection
**Description:**

A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.

**Requirements:**

- Identify duplicate user IDs.
- Display how many times each duplicate appears.


```python
## Taking number of users from the user
no_of_users = int(input("Enter number of users: "))

## Empty list to store user id's

user_id = []

for i in range(no_of_users):
    user_id.append(input("Enter user_id: "))

unique_id = set(user_id)

## Logic to detect duplicate users

for i in unique_id:
    if user_id.count(i) > 1:
        print(f"{i} --> {user_id.count(i)} times")
```
