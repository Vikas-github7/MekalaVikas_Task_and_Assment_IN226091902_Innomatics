## Problem Statement 1:Smart Parking Lot Management System
Design a function to manage a smart parking lot.

**The system should:**
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity


```python
## Librarey to evaluate user input safely
import ast

## Smart Parking Lot Management System
def parking_lot_management():
    ## Taking parking capacity from the user
    capacity = int(input("Parking Capacity: "))
    ## Taking Vehicle logs from the user
    log = ast.literal_eval(input("Vehicle Logs: "))

    ## Counter that calculates car lots being used
    count = 0
    ## Calculating the number of parked cars
    for i in log:
        if i == "IN":
            count += 1
        else:
            count -= 1
    parked = count

    ## Calculating the lot available status
    if parked < capacity:
        status = "Available"
    else:
        status = "Not Available"
    return parked, status

parked_cars, parking_status = parking_lot_management()

print(f"Currently Parked Vehicles: {parked_cars}")
print(f"Parking Status: {parking_status}")
```

## Problem Statement 2:Online Food Delivery Time Estimator
Create a function that estimates **delivery time** based on:
- Distance (km)
- Weather condition
- Traffic level

Apply delays dynamically and display final ETA.


```python
## Online Food Delivery Time Estimator
def food_delivery_time_estimator():

    ## Taking distance from the user
    distance = int(input("Distance (km): "))
    ## Taking Traffic level  from the user
    traffic = input("Traffic Level: ").lower()
    ## Taking weather status from the user
    weather = input("Weather: ").lower()

    ## Lets assume for each 1 km it takes 5 minutes time
    time = distance * 5

    ## If the traffic level is medium
    if traffic == "medium":
        time = 1.1 * time
    ## If the traffic level is high
    if traffic == "high":
        time *= 1.2
    ## If the weather is rainy
    if weather == "rainy":
        time *= 1.5
    return time

Time = food_delivery_time_estimator()

print(f"Estimated Delivery Time: {Time:.0f} minutes")
```

## Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
Build a function that analyzes seat booking data and:
1. Calculates occupancy percentage
2. Determines if show is Housefull
3. Suggests opening additional shows


```python
## Librarey used to evaluate user input safely
import ast

## Movie Theatre Seat Occupancy Analyzer
def seat_occupancy():
    ## Taking Total Seats of the theatre from the user
    seats = int(input("Total Seats: "))

    ## For every booked seat 1 in list
    Booked = ast.literal_eval(input("Booked Seats:"))
    booked_seats = 0

    ## Calculating Total booked seats
    for i in Booked:
        booked_seats += i

    ## Calculating the occupancy of the theatre
    occupancy = booked_seats/seats * 100

    ## Logic to determine the status of the show
    if occupancy < 50:
        return occupancy, "Fast Filling"
    elif occupancy == 100:
        return occupancy, "Full"
    elif occupancy >= 75:
        return occupancy, "Almost Full"
    else:
        return occupancy, "Half Full"

occupancy, status = seat_occupancy()

print(f"Occupancy: {occupancy:.0f}%")
print(f"Show Status: {status}")
```

## Problem Statement 4:Cloud Server Load Classification System
Create a function to classify **server load** based on CPU usage readings.

**Rules:**
- Average CPU < 50% → Normal
- 50%–80% → Warning
- 80% → Critical


```python
## Librarey used to evaluate user input safely
import ast
## Cloud Server Load Classification System
def server_load_classifier():
    ## Taking CPU readings from the user
    readings = ast.literal_eval(input("CPU Readings (%): "))
    load = 0

    ## Calculating total load
    for i in readings:
        load += i

    ## Calculating average load
    avg = load/len(readings)

    ## Determing load status based on average CPU load
    if avg >= 80:
        return avg, "Critical"
    elif 50<=avg<80:
        return avg, "Warning"
    else:
        return avg, "Normal"


average, status = server_load_classifier()
print(f"Average CPU Load: {average}")
print(f"Server Status: {status}")
```

## Problem Statement 5:Smart Classroom Resource Usage Monitor
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies **overuse patterns**.

**Real-Time Use**
- Smart classrooms
- Energy optimization systems

```python
## Librarey used to evaluate user input safely
import ast

## Smart Classroom Resource Usage Monitor
def resource_usage_monitor():

    ## Taking dictionary resource: hours used from the user
    resource_usage = ast.literal_eval(input("Resource Usage (hours): "))

    ## Threshold energy hpirs per each resource 
    max_hours = 6
    ## Empty list to store Over used resources
    over_usage = []
    ## Setting the alert as "no"
    alert = "No"

    ## Logic to get over used resources and alert
    for i in resource_usage:
        if resource_usage[i] <= max_hours:
            continue
        else:
            over_usage.append(i)
            alert = "Yes"
    return over_usage, alert

over_used, energy_alert = resource_usage_monitor()

print(f"Overused Resources: {", ".join(over_used)}")
print(f"Energy Alert: {energy_alert}")
```

## Problem Statement 6: Online Event Registration Capacity Controller
Create a function that manages event registrations by:
1. Tracking registrations
2. Preventing overbooking
3. Triggering waitlist mode


```python

## Online Event Registration Capacity Controller
def event_registration_management():
    ## Taking the event capacity from the user
    capacity = int(input("Event Capacity: "))
    ## Taking number of registrations from the user
    registred = int(input("Registrations: "))

    ## Initilizing the variables
    waitlist = 0
    status = "Open"

    ## Logic to determine status and calculate confirmed registrations and waitlisted users
    if registred >= capacity:
        confirmed = capacity
        waitlist = registred - capacity
        status = "Closed"
        return confirmed, waitlist, status
    else:
        return registred, waitlist, status

confirm, wait, status = event_registration_management()
print(f"Confirmed Registrations: {confirm}")
print(f"Waitlisted Users: {wait}")
print(f"Registration Status: {status}")
```
