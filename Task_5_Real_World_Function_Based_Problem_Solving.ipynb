## Problem 1: Social Media – Post Engagement Analyzer
**Problem Statement**

Create a function to analyze engagement on a social media post.

**Rules**
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement


```python
## Librarey used to evaluate user input safely
import ast

## List of like for the post
likes = ast.literal_eval(input("Enter likes data: "))

## Post Engagement Analyzer Function
def post_analyze(likes):
    ## Initiliaze total likes
    total_likes = 0

    ## Logic to calculate total likes
    for i in likes:
        total_likes += i

    ## Logic to check Post Engagement on social media
    if total_likes >= 1000:
        return total_likes, "Viral Post"
    else:
        return total_likes, "Normal Engagement"

total, status = post_analyze(likes)

print(f"Total Likes: {total}")
print(f"Post Status: {status}")
```

## Problem 3: Agriculture – Rainfall Adequacy Checker
**Problem Statement**

Create a function that analyzes seasonal rainfall data.

**Rules**
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall


```python
## Library used to evaluate input safely from the user
import ast

## Taking the list of rainfall values from the user
rainfall_list = ast.literal_eval(input("Enter the rainfall values: "))

## Rainfall Adequacy Checker Function
def rainfall_checker(rain_fall):
    ## Initilizing required rainfall
    required_rain_fall = 50

    ## Calculating Average Rainfall
    avg_rain_fall = sum(rain_fall)/len(rain_fall)

    ## Logic to check Adequacy of the rainfall
    if avg_rain_fall >= required_rain_fall:
        return avg_rain_fall, "Adequate Rainfall"
    else:
        return avg_rain_fall, "Inadequate Rainfall"

avg, status = rainfall_checker(rainfall_list)

print(f"Average Rainfall: {avg}")
print(f"Rainfall Status: {status}")
```

## Problem 4: Social Media – Duplicate Account Detection
**Problem Statement**

Create a function to detect duplicate usernames in a system.

**Rules**
- If duplicates exist → Flag accounts
- Else → Accounts are unique


```python
## Taking number user data from the user
number_of_users = int(input("Enter the number of users: "))
## Empty list to store user names
user_names = []
## Taking list of usernames from the user
for i in range(number_of_users):
    user_names.append(input("Enter user name: "))

## Function to detect duplicate accounts
def duplicate_user(names):
    unique_names = set(names)
    if len(unique_names) == len(names):
        return "No"
    else:
        return "Yes"

is_duplicate = duplicate_user(user_names)
print(f"Duplicate Accounts Found: {is_duplicate}")
```

## Problem 6: Agriculture – Premium Crop Price Filter
**Problem Statement**

Create a function that filters crop prices above a premium threshold.

**Rules**
- Price > ₹2000 → Premium Crop
- Else → Regular Crop


```python

## Taking the number of crops from the user
number_of_crops = int(input("Enter the number of crops: "))
## Empty list to store crop prices
crop_prices = []

## Logic to get crop prices from the user
for i in range(number_of_crops):
    crop_prices.append(int(input("Enter the crop price: ")))

## Premium Crop Price Filter
def price_filter(crops):
    ## Empty list to store premium crop prices
    premium_crops = []

    ## Logic to detect premium crops
    for i in crops:
        if i > 2000:
            premium_crops.append(i)
    return premium_crops

premium = price_filter(crop_prices)

print(f"Premium Crops: {premium}")
            
```

## Problem 7: System Monitoring – Application Health Checker
**Problem Statement**

Create a function to evaluate application health based on error count.

**Rules**
- Errors = 0 → Healthy
- Errors ≤ 5 → Minor Issues
- Errors > 5 → Critical Issues


```python

## Function to check Application Health
def app_health_checker():
    ## Taking number of errors from the user
    errors = int(input("Enter number of Errors: "))

    ## Logic to determine Application Health
    if errors == 0:
        return errors,"Healthy"
    elif errors <= 5:
        return errors,"Minor Issues"
    else:
        return errors,"Critical Issues"

## Calling the function and storing the return values
error,app_health = app_health_checker()
print(f"Error Count: {error}")
print(f"System Status: {app_health}")
```

## Problem 8: Banking – Daily Transaction Limit Checker
**Problem Statement**

Create a function to validate daily transaction amount.

**Rules**
- Daily limit = ₹50,000
- Amount within limit → Approved
- Amount exceeds limit → Rejected


```python
## Function to limit daily transaction
def transaction_limit_checker():
    ## Taking transaction amount from the user
    amount = int(input("Enter the amount: "))
    ## Daily limit
    daily_limit = 50000
    ## Logic to verify transaction
    if amount <= daily_limit:
        return amount, "Approved"
    else:
        return amount, "Rejected"

## Calling the function and storing return values
amount, status = transaction_limit_checker()

print(f"Transaction Amount: {amount}")
print(f"Transaction Status: {status}")
```

## Problem 9: E-Learning – Student Attendance Eligibility System
**Problem Statement**

Create a function to calculate attendance percentage and check exam eligibility.

**Rules**
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible


```python
## Library used to evaluate user input safely
import ast

## Exam Eligibility System
def exam_eligibility():

    ## Taking the attendance from the user
    ## 1 == "Present"
    ## 0 == "Absent"
    attendance = ast.literal_eval(input("Enter the attendance: "))

    ## Calculating the attendance percentage
    attend_perc = sum(attendance)/len(attendance) * 100

    ## Checking Exam Eligibility
    if attend_perc >= 75:
        return attend_perc, "Eligible"
    else:
        return attend_perc, "Not Eligible"

## Calling the function and storing return values
perc, eligibility = exam_eligibility()

print(f"Attendance Percentage: {perc}")
print(f"Exam Eligibility: {eligibility}")
```

## Problem 10: Smart Electricity Bill Analyzer
**Problem Statement**

You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules**
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit

**Usage Classification**
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage


```python

## Electricity Bill Analyzer
def bill_analyzer():

    ## Units consumed by the user
    units = int(input("Enter Total Units Consumed: "))

    ## Initilizing the total bill
    cost = 0

    ## Calculating the total bill
    for i in range(1, units+1):
        if i <= 100:
            cost += 3
        elif i <= 200:
            cost += 5
        else:
            cost += 7

    ## Checking the usage
    if cost < 500:
        return cost, "Low Usage"
    elif cost <= 1500:
        return cost, "Moderate Usage"
    else:
        return cost, "High Usage"

## Calling the bill_analyzer function and storing the values
bill, usage = bill_analyzer()

print(f"Bill: {bill}")
print(f"Usage Status: {usage}")
```
