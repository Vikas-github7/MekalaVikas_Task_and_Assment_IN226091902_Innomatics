## Problem 1: Unique Words in a Sentence

**Problem Statement**

Given a sentence, identify all unique words and display:

1. The count of unique words
2. The unique words themselves


```python
## Taking input from the user
sentence  = input("Enter The sentence : ")

## Converting the string to list
sen_list = sentence.split()

## Covert the list into set, to avoid duplicates

## To find the count of unique words
print(f"The ount of unique words : {len(set(sen_list))}")

## To find the unique words
print(f"Unique words : {set(sen_list)}")



```

## Problem 2: Highest Salary from Employee Data
**Problem Statement**

Store employee names and their salaries, and determine which employee has the highest salary.


```python

## Taking the input from the user as dictionary

emp_data = eval(input())
## eval() ---> evaluates the input

## Initilizing the variables
max_sal = 0
name = ""

## Logic to find Highest salary and the Highest salary employee
for key in emp_data:
    if emp_data[key] > max_sal:
        max_sal = emp_data[key]
        name = key

print(f"Highest Salary : {name} - {max_sal}")
```

## Problem 3: Find Maximum and Minimum Values
**Problem Statement**

Find the maximum and minimum values from a list of numbers.


```python

## Taking the input from the user as a list of numbers

## eval() --> evaluates the input
num = eval(input("Enter list of numbers: "))

## Setting the max and min values
max_num = num[0]
min_num = num[0]

## Logic to find Maximum and Minimum Values

for i in num:
    ## To find maximum
    if i  > max_num:
        max_num = i
    ## To find minimum
    if i < min_num:
        min_num = i

print(f"List: {num}")
print(f"Max: {max_num}")
print(f"Min: {min_num}")
```

## Problem 5: Calculate Attendance Percentage
**Problem Statement**

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
## Taking the input from the user as a List

attendance = eval(input("Enter the attendance: "))
## eval() --> evaluates the input entered by the user

## Initilizing the variables
present = 0
no_of_days = len(attendance)

## Logic to find the Attendance Percentage

for i in attendance:
    if i == "P":
        present += 1

perc = present/no_of_days*100

print(f"Attendance Percentage: {perc}")
```

## Problem 7: Count Character Frequency
**Problem Statement**

Count how many times each character appears in a given string.


```python

## Taking the input from the user as a string
string = input("Enter a string: ")

## Converting the string into a list
string_list = list(string)

## Removing the duplicates from the string list
unique_characters = set(string_list)

## Initilizing an empty dictionary
dict1 = {}

## Logic to count the occurances of each characters
for i in unique_characters:
    dict1[i] = string.count(i)

print(dict1)
```

## Problem 9: Check if a Key Exists in a Dictionary
**Problem Statement**

Check whether a specific key exists in a dictionary.


```python

## Dictionary containing employee id and employee name
employee = {1: "Vikas", 2: "Kowshik", 3: "Megha", 4: "Priya", 5: "Akansha"}

## Getting the employee id from the user
emp_id = int(input("Enter the employee id: "))

## Check whether the employee exists in the dictionary using "in"
## in ---> membership operator

if emp_id in employee:
    print("The employee exists")
    print("Key exists in the dictionary")
else:
    print("The employee doesn't exist")
    print("Key doesnot exist in the dictionary")
```

## Problem 10: Calculate Average Marks
**Problem Statement**

Calculate the average marks of a student from a list of marks.


```python
## Taking the input from the user as a list of marks

marks = eval(input("Enter the marks of an student: "))
## eval() --> evaluates the input

## To find total marks
total_marks = sum(marks)

## To find the total number of subjects
subjects = len(marks)

## Calculating the average of the student

average = total_marks/subjects

print(f"Average Marks: {average}")
```
